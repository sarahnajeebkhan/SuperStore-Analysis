# Data Wrangling & Analysis

## 1> Trend of Sales and Profit annually

```python
sstore_data['Order Date'] = pd.to_datetime(sstore_data['Order Date'])
sstore_data['Year']=sstore_data['Order Date'].dt.year
sstore_data['Year'].head()
```

```python
sstore_data.groupby('Year')['Sales','Profit'].agg(['sum']).plot.bar()
plt.title('Annual trend of Sales and Profit')
```

### We observe a Linear Trend in the Profits generated by the SuperStore annually although the Sales experienced a slight dip in the year2015

## 2> What is the distribution of orders places in each Sub-Category?

```python
plt.figure(figsize=(12,10))
sstore_data['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

## 3> Which Sub-Category experiences the maximum Sales and which Sub-Category gebetates the maximum Profit?

```python
plt.rcParams["figure.figsize"] = (10,5)
sstore_data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
# plt.legend('Profit')
# plt.legend('Sales')
plt.show()
```

### We observe that :-
    1) The subcategories "Chairs" and "Phones" generate the maximum Sales
    2) The subcategory "Copiers" generate the maximum Profit
    3) The subcategories "Bookcases" and "Tables" are generating Loses

```python
sstore_data['Cost']=sstore_data['Sales']-sstore_data['Profit']
```

```python
sstore_data['Profit%']=(sstore_data['Profit']/sstore_data['Cost'])*100
```

```python
sstore_data.sort_values(['Profit%','Product Name'],ascending=False)
```

## 4>Who are the top 10 most profitable customers of the SuperStore?

```python
Top10 = sstore_data.sort_values(['Profit'], ascending=False).head(10)
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
p = sns.barplot(x='Customer Name', y='Profit',hue='State',palette='Set1', data=Top10, ax=ax)
ax.set_title("Top 10 most profitable Customers")
ax.set_xticklabels(p.get_xticklabels(), rotation=75)
plt.tight_layout()
plt.show()
```

## 5> Which Shipping Mode do Customers choose based on the Category and Quantity?

```python
sstore_shipmode = sstore_data[['Ship Mode','Category','Quantity']].groupby(['Ship Mode','Category']).sum('Quantity')
sstore_shipmode
```

```python
sstore_shipmode.plot(kind='bar', title='Ship Mode vs Sum of Quantity', ylabel='Sum of Quantity',
         xlabel='Ship Mode with Category', figsize=(10, 5))
```

### We observe that  "Standard Class" is the most used Shipping Mode while "Same Day" is the least used Shipping Mode used by Customers

## 6> Which are the Top 10 most selling Products of the SuperStore?

```python
sstore_product = pd.pivot_table(data=sstore_data[['Product Name','Quantity']], 
                        index='Product Name', 
                        values='Quantity',
                        aggfunc='sum')
top10_product=sstore_product.sort_values('Quantity',ascending=False)
top10_product.head(10)
```

```python
top10_product.head(10).plot(kind='bar', title='Bestselling Products vs their Quantity', ylabel='Quantity', 
                            xlabel='Product Name', figsize=(10, 5))
```

### Observation : Staples and staple related products are the most selling products of the SuperStore

## 7>Which state generates the Highest Sales for the SuperStore?

```python
sstore_state_sales = pd.pivot_table(data=sstore_data[['State','Sales']], 
                        index=['State'], 
                        values='Sales',
                        aggfunc='sum')
sstore_state_sales.sort_values(by='Sales', ascending=False).head(10)
```

```python
sstore_state_sales.plot(kind='bar', title='State vs Maximum Sales', ylabel='Maximum Sales',
         xlabel='State', figsize=(10, 5));
```

### Observation : "California" generates the Highest Sales

## 8> Which State generates the Maximum Sales for the SuperStore?

```python
sstore_max_sales = pd.pivot_table(data=sstore_data[['State','Sales']], 
                        index=['State'], 
                        values='Sales',
                        aggfunc='max')
sstore_max_sales.sort_values(by='Sales', ascending=False).head(10)
```

```python
sstore_max_sales.plot(kind='bar', title='Maximum Sales per Sate', ylabel='Maximum Sales',
         xlabel='State', figsize=(10, 5));
```

### Observation : "Florida" generates the Maximum Sales

## 9> Which State generates the Highest Profit for the SuperStore?

```python
sstore_state_profit = pd.pivot_table(data=sstore_data[['State','Profit']], 
                        index=['State'], 
                        values='Profit',
                        aggfunc='sum')
sstore_state_profit.sort_values(by='Profit', ascending=False).head(10)
```

```python
sstore_state_profit.plot(kind='bar', title='State vs Highest Profit', ylabel='Profit',
         xlabel='State', figsize=(10, 5));
```

### Observation : "California" generates the Highest Profit
